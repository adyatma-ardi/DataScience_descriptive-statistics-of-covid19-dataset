```python
import pandas as pd
```

```python
# membaca file csv
df = pd.read_csv('covid_19_indonesia.csv')

# menampilkan dataframe
df.head(5)
```

```python
# menampilkan dataframe
df.tail(5)
```

```python
df.shape # menampilkan jumlah baris dan kolom
df.info() # menampilkan informasi tipe data
df.describe() # menampilkan statistik deskriptif
```

```python
print('mean :', df.mean())  # Menghitung mean (rata-rata)
print('median :', df.median())  # Menghitung median
print('modus: ', df.mode())  # Menghitung modus
print('min :', df.min())  # Menghitung nilai minimum
print('max :', df.max())  # Menghitung nilai maksimum
print('standard deviation :', df.std())  # Menghitung standar deviasi
```

```python
import matplotlib.pyplot as plt

df[['New Cases', 'New Deaths', 'New Cases per Million' ,'New Deaths per Million']].hist(color='black') # menampilkan histogram
```

```python
df.boxplot(rot=90) # menampilkan boxplot
```

```python
df.plot.scatter(x='New Recovered', y='New Deaths', color='black') #menampilkan diagram scatter plot
```

```python
df.isnull().sum() # menghitung jumlah nilai hilang pada setiap kolom
df.dropna() # menghapus baris yang mengandung nilai hilang
df.fillna(0) #mengisi nilai hilang dengan nilai tertentu
```
